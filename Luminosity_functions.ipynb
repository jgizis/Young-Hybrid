```python
def mass_powerlaw(size,m0,m1,alpha):
    #dN/dM ~ M**(-alpha)
    factor1= (m1**(1-alpha)-m0**(1-alpha))
    factor2 = m0**(1-alpha)
    return (factor1*np.random.rand(size)+factor2)**(1/(1-alpha))
```

```python
#Let us make a sample drawn from a power-law mass distribution...
sample_size=10000
star_masses=mass_powerlaw(sample_size,0.002,0.03,0.6)
```

```python
#... and a uniform distribution of ages
# from 5 Myr to 250 Myr
star_ages = np.random.uniform(low=5.0,high=250.0,size=sample_size)
```

```python
plt.plot(star_ages,star_masses,'.',alpha=0.1)
```

```python
plt.hist(star_masses)
```

```python
plt.hist(star_ages)
```

```python
star_logTeff=CELLO_interp_logTeff(star_masses,star_ages)
```

```python
plt.hist(10.0**star_logTeff,range=[500,1800],bins=13)
plt.xlim(500,1500)
```

```python
plt.plot(star_ages,10.0**star_logTeff,'.',alpha=0.10)
plt.plot(model030hot['col3']/1e6,10.0**model030hot['col11'],'k')
plt.plot(model002hot['col3']/1e6,10.0**model002hot['col11'],'k')
plt.xlim(5,250)
plt.xlabel('Age (Myr)')
```

The hybrid model creates an excess at 1100-1200K. Note, however, the lack of points for ~80-100Myr.  This is because the D- burning models (0.012 and higher) take 100Myr to reach the transition but the non-D-burning (0.011 and lower) reach it in less than 80 Myr.
